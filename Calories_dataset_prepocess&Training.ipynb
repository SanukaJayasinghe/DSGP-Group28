```python
import pandas as pd
import xgboost as xgbregressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
features = ['age', 'weight(kg)', 'height(m)', 'BMI', 'activity_level','le_gender']
#features = ['le_age', 'le_weight','le_gender' , 'le_height', 'le_bmi', 'le_bmr', 'le_a_level']
target = 'calories_to_maintain_weight'
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1[features], df1[target], test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
from xgboost import XGBRegressor

# Instantiate the XGBRegressor
model = XGBRegressor(objective="reg:squarederror", n_estimators=1000)

# Fit the model to your data
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
X_test
```

```python
import matplotlib.pyplot as plt
x_ax = range(len(y_test[500:600]))
plt.plot(x_ax, y_test[500:600], label="original")
plt.plot(x_ax, y_pred[500:600], label="predicted")
plt.title("Boston test and predicted data")
plt.legend()
plt.show()
```

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate regression evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print(f"Root Mean Squared Error: {rmse:.2f}")
print("R-squared:", r2)
print(f"R-squared: {r2:.2f}")
```
